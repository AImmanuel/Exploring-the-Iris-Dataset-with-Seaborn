```python
from sklearn.datasets import load_iris
```

```python
dataframe = load_iris()
dataframe
```

```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.DataFrame(data= np.c_[dataframe['data'], 
                                     dataframe['target']],
                  columns= dataframe['feature_names'] + ['Species'])
dataframe
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("dark")
```

## 1. Scatterplot

A scatter plot is a type of plot used to display values for typically two variablesin a dataset. 
Sepal length v Sepal Width.

```python
sns.lmplot(x='sepal length (cm)', 
           y='sepal width (cm)', 
           data=dataframe)
```

```python
sns.lmplot(x='sepal length (cm)', 
           y='sepal width (cm)', 
           data=dataframe,
           fit_reg=False, # remove regression line
           hue='Species')   # set color by species
```

## 2. Boxplot 

A boxplot is a standardized way of displaying the distribution of data based on a five number summary
- 1.Minimum           
- 2.1st quartile(Q1)
- 3.Median            
- 4.3rd quartile(Q3)
- 5.Maximum
- It also provides information about outliers and their values.

```python
plt.figure(figsize=(10,8)) 
sns.boxplot(data=dataframe)
```

```python
dataframe_1 = dataframe.drop(['Species'], axis=1) # Remove Species column

plt.figure(figsize=(10,8)) 
sns.boxplot(data=dataframe_1)
```

```python
species_colors = ['#78C850',  # Setosa
                    '#F08030',  # Versicolor
                    '#6890F0',  # Verginica
                  ]
```

## 3. Violin Plot

A violin plot is used to visualize the distribution of numerical data. Violin plots depict summary statistics and the density of each variable. 

```python
plt.figure(figsize=(10,8)) 
sns.violinplot( data=dataframe, 
               palette=species_colors) 
```

## 4. Swarm Plot

A swarm plot is another way of plotting the distribution of an attribute or the joint distribution of a couple of attributes. 
Here we create a Swarm plot of Petal length across the species.

```python
plt.figure(figsize=(10,8)) 
sns.swarmplot(y='petal length (cm)',
              x='Species', 
              data=dataframe, 
              palette=species_colors)
```

```python
# Combining the Violin and Swarm plots
plt.figure(figsize=(10,8))
 
sns.violinplot(x='Species',
               y='petal length (cm)', 
               data=dataframe, 
               inner=None, # Remove the inner bars of the violin plots
               palette=species_colors)
 
sns.swarmplot(y='petal length (cm)', 
              x='Species', 
              data=dataframe, 
              color='w', # Make points white
              alpha=0.5) # Transparency
 
plt.title('Combined Plot')
```

## 5. Correlation Heatmap

Correlation describes the size and direction of a relationship between two (or more) variables. i.e. the directionality of the variables.

```python
corr = dataframe_1.corr()
corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="crest")
```

## 6. Histogram

```python
plt.figure(figsize=(10,8))
sns.histplot(dataframe,
             x='petal length (cm)',
             kde=True)
```

## 7. Bar Plot

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Species',
              data=dataframe, 
              palette=species_colors)
plt.xticks(rotation=-45) 
```

## 8. CatPlot

```python
g = sns.catplot(x='Species', 
                   y='petal width (cm)', 
                   data=dataframe, 
                   hue='Species',  # Color by species
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
```

## 9. Density Plot

Density plot represents the distribution of a numeric variable that uses a kernel density estimate to show the probability density function of the variable.

```python
sns.kdeplot(x='sepal length (cm)',
            y='sepal width (cm)',
            data=dataframe)
```

## 10. Joint Distribution Plot

Joint distribution plots combine information from scatter plots and histograms for bi-variate distributions. We also join a Density plot onto it.

```python
h= sns.jointplot(x='petal length (cm)',
              y='petal width (cm)',
              hue="Species",
              data=dataframe)

h.plot_joint(sns.kdeplot, 
             color="r", 
             zorder=0, 
             levels=6)
```
